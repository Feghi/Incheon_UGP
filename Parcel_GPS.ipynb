```python
import os
import time
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
from pyarrow import csv
from tqdm import tqdm 
import sys
pd.options.display.max_columns = 500
```

```python
path_dir='E:/parcel'
file_list = os.listdir(path_dir+"/")
```

```python
len(file_list)
```

```python
post = pd.read_csv('E:/post/picmt0060.csv', encoding='UTF-8', sep=",", error_bad_lines=False)
subpost=post[post['POADDR'].str.contains('인천')]
subpost = subpost.iloc[:,[0,16,18,19]]
```

```python
subpost.columns = ['REGI_POCD','PO_NM','ZIP_CD','PO_ADDR']
subpost['REGI_POCD']=subpost['REGI_POCD'].astype(str)
subpost['ZIP_CD']=subpost['ZIP_CD'].astype(str)
subpost['ZIP_CD'] = subpost['ZIP_CD'].apply(
    lambda x: 
        x[0:5])
subpost.head()
```

```python
for i in tqdm(file_list):
    df = pd.read_csv(path_dir+"/"+i, encoding='UTF-8', sep="|", error_bad_lines=False)
    df_sample = df.iloc[:,[ 0,5,7,8,36,108,109]]
    df_sample.columns = ['REGI_NO','REC_ADDR_SEQ','WGHT','VOL','REGI_POCD','경도','위도']
    df_sample['REGI_POCD'] = df_sample['REGI_POCD'].astype(str)
    df_sample['REGI_POCD'] = df_sample['REGI_POCD'].apply(
    lambda x: 
        x[0:5])
    df_sample=subpost.merge(df_sample, on='REGI_POCD', how='left')
    df_sample.to_csv('E:/incheon/parcel_'+i[10:18]+'_'+str(df_sample.shape[0])+'.csv', mode='a', sep="|")  
```
