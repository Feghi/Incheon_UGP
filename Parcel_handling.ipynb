```python
import os
import time
import pandas as pd
from tqdm import tqdm 
import sys
pd.options.display.max_columns = 500
```

```python
path_dir='E:/incheon/'
file_list = os.listdir(path_dir)
```

```python
temp = pd.DataFrame()
for i in tqdm(file_list):
    df = pd.read_csv(path_dir+i, encoding='UTF-8', sep="|", error_bad_lines=False)
    df['YMD']=i[7:15]
    df=df.dropna(axis=0)
    temp = pd.concat([df,temp])
temp.to_csv('E:/incheon/parcel_'+str(temp.shape[0])+'.csv', mode='a', sep="|")  
```
